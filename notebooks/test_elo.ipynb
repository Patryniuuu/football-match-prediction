```python
from scraping.transform import add_elo_rating
x = add_elo_rating(matchday_df)

x
```

```python
x["dni_miedzy_meczami"] = x["Date"].diff().dt.days
x
```

```python
x["elo_diff"] = (x["Elo_Team"] - x["Elo_Opponent"]).round(2)
x
```

```python
x.drop("Time",axis=1)
```

```python
x["avg_elo_of_5_last_opp"] = x["Elo_Opponent"].rolling(window=5,min_periods=1).mean()
x
```

```python

y = x.copy()
y["GF"] = pd.to_numeric(y["GF"], errors="coerce")
y["GA"] = pd.to_numeric(y["GA"], errors="coerce")
y["xG"] = pd.to_numeric(y["xG"], errors="coerce")
y["xGA"] = pd.to_numeric(y["xGA"], errors="coerce")
y["goal_diff"] = y["GF"] - y["GA"]
y["exp_goal_diff"] = (y["xG"] - y["xGA"]).round(2)
y["avg_goals_scored_in_5_games"] = y["GF"].rolling(window=5,min_periods=1).mean().round(2)
y["av_goals_conceded_in_5_games"] = y["GA"].rolling(window=5,min_periods=1).mean().round(2)

y

```

```python
y["Venue_code"] = y["Venue"].astype("category").cat.codes
y["Result_code"] = y["Result"].astype("category").cat.rename_categories({"W":2,"D":1,"L":0})
y.columns
```

```python
y["last_5_opp_avg_elo_exluding_present_match"] = y["Elo_Opponent"].shift(1).rolling(window=5, min_periods=1).mean()
y[["Elo_Opponent", "last_5_opp_avg_elo_exluding_present_match"]]
```

```python
from config import FBREF_ABR_TO_FBREF
# teraz sprobujemy polaczyc mecze z obu stron w jeden rekord
#zaczne od odfiltrowania meczow tylko z PL
df = y.loc[y["Comp"] == "Premier League", :]
df["Opponent"] = df["Opponent"].apply(lambda x: FBREF_ABR_TO_FBREF.get(x))
df
```

```python
#teraz muszę utworzyć df_home i df_away
df_home = df.loc[df["Venue"] == "Home"].copy()
df_home = df_home.rename(columns=lambda x: f"{x}_H" if x not in ['Date', 'Team', 'Opponent', 'Time', 'Comp','Round','Season'] else x)
df_away = df.loc[df["Venue"] == "Away"].copy()
df_away = df_away.rename(columns=lambda x: f"{x}_A" if x not in ['Date', 'Team', 'Opponent', 'Time', 'Comp','Round', 'Season'] else x)
df_away
```

```python
# Klucz do łączenia: data + para drużyn
df_home['match_key'] = df_home['Date'].astype(str) + "_" + df_home['Team'] + "_" + df_home['Opponent']
df_away['match_key'] = df_away['Date'].astype(str) + "_" + df_away['Opponent'] + "_" + df_away['Team']

dup_columns = [col for col in df_home.columns if col in df_away.columns]
dup_columns = dup_columns[:-1]
dup_columns
df_away = df_away.drop(dup_columns, axis=1)

#Merge
df_merged = pd.merge(df_home, df_away, on='match_key', suffixes=('', ''))  # brak sufiksów, bo już mamy _H / _A
df_merged

```

```python
# Dodaj czysto: HomeTeam i AwayTeam
df_merged['HomeTeam'] = df_merged['Team']
df_merged['AwayTeam'] = df_merged['Opponent']

df_merged
```
